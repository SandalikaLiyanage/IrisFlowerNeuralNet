```python
import pandas as pd
dataset=pd.read_csv(r"C:\Users\Sandalika\Documents\deep learning workshop\iris.csv").values
print(dataset, dataset.shape)
```

```python
data=dataset[:,:4]
target=dataset[:,4]
print(data.shape,target.shape)
```

```python
print(data)
print(target)
```

```python

from keras.utils import to_categorical
new_target=to_categorical(target)
print(target[:10])
```

```python
print(new_target[:10])
```

```python
from sklearn.model_selection import train_test_split
train_data, test_data,train_target,test_target=train_test_split(data,new_target,test_size=0.1)
```

```python
print(train_data,train_data.shape)
```

```python
print(test_data,test_data.shape)
```

```python
from keras.models import Sequential
from keras.layers import Dense

model=Sequential() #empty sequential nn
model.add(Dense(64,input_dim=4,activation='relu')) #first hidden layer
model.add(Dense(32,activation='relu'))#2nd hudden layer
model.add(Dense(3,activation="softmax")) #output layer
model.compile(loss="categorical_crossentropy",optimizer='adaGrad',metrics=['accuracy'])
```

```python
model.summary()
```

```python
for layer in model.layers:
        print("===================================",layer.name,'======================================')
        parameters=layer.get_weights()
        print(parameters)
```

```python
#training then model
model.fit(train_data,train_target,epochs=200,validation_split=0.1)
```

```python
from matplotlib import pyplot as plt
plt.plot(model.history.history['loss'], label='loss')
plt.plot(model.history.history['val_loss'], label='val loss')
plt.legend()
plt.xlabel('#epoch')
plt.ylabel('loss')
plt.show()
```

```python
plt.plot(model.history.history['accuracy'])
plt.plot(model.history.history['val_accuracy'],label='val accuracy')
plt.legend()
plt.xlabel('#epoch')
plt.ylabel('accuracy')
plt.show()
```

```python
import numpy as np
test_data=np.array([5.3,2.3,4.7,2.1]).reshape(1,-1)
predicted_target=model.predict(test_data)
print(predicted_target)
```
